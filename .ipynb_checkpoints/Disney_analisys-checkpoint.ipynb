```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
disney_full = pd.read_csv('DIS.csv')
```

```python
disney_full.head()
```

```python
disney = disney_full[['Date', 'Open']]
disney.Date = pd.to_datetime(disney.Date)
disney['Year'] = pd.DatetimeIndex(disney['Date']).year
disney.head()
```

```python
plt.figure(figsize=(13, 10))
plt.plot(disney.Date, disney.Open)
plt.title('Ежедневные данные')
plt.grid();
```

```python
#disney['Year'] = pd.DatetimeIndex(disney['Date']).year
disney_by_year = disney.groupby('Year', as_index=False).agg({'Open': 'mean'})
#disney_by_year.Year = pd.to_datetime(disney_by_year.Year, format="%Y")
disney_by_year.head()
```

```python
plt.figure(figsize=(13, 10))
sns.lineplot(x=disney_by_year.Year, y=disney_by_year.Open)
plt.grid()
plt.title('Среднее значение по годам')
plt.yticks(np.arange(0, max(disney_by_year.Open) + 10, 10));
```

```python
plt.figure(figsize=(12, 10))
plt.semilogy(disney.Date, disney.Open, base=np.exp(1))
plt.title('Полулогарифмический график')
plt.grid()

labels = np.arange(6)
points = np.exp(labels)
plt.yticks(points, labels);
```

```python
#fig, ax = plt.subplots()

plt.figure(figsize=(12, 10))

plt.semilogy(disney.Date, disney.Open, base=np.exp(1))
plt.title('Полулогарифмический график с трендами')
plt.grid()

labels = np.arange(6)
points = np.exp(labels)
plt.yticks(points, labels)


z1 = np.polyfit(disney.Year[disney.Year < 2003], disney.Open[disney.Year < 2003], 1)
p1 = np.poly1d(z1)
plt.plot(disney.Date[disney.Year < 2003], p1(disney.Year[disney.Year < 2003]), "r")

z2 = np.polyfit(disney.Year[disney.Year >= 2003], disney.Open[disney.Year >= 2003], 1)
p2 = np.poly1d(z2)
plt.plot(disney.Date[disney.Year >= 2003], p2(disney.Year[disney.Year >= 2003]), "r")


plt.show();
```
